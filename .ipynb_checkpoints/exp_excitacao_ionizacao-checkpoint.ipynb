```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Carregar os dados do arquivo CSV
df = pd.read_csv("dados_organizados.csv")

# Renomear colunas para facilitar
df = df.rename(columns={"V": "I", "corrente": "VCG"})

# Garantir que os dados são numéricos
df["VCG"] = pd.to_numeric(df["VCG"], errors='coerce')
df["I"] = pd.to_numeric(df["I"], errors='coerce')
df = df.dropna(subset=["VCG", "I"])

# Detectar picos SEM definir altura mínima
peaks_indices, _ = find_peaks(df["I"])

# Obter os valores correspondentes aos picos
vcg_peaks = df["VCG"].iloc[peaks_indices]
i_peaks = df["I"].iloc[peaks_indices]

# Verificar se algum pico foi encontrado
if len(vcg_peaks) > 0:
    print("Picos encontrados:")
    for v, i in zip(vcg_peaks, i_peaks):
        print(f"VCG = {v:.2f} V, I = {i:.2f} μA")
else:
    print("Nenhum pico foi encontrado.")

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df["VCG"], df["I"], marker='o', linestyle='-', color='purple')
plt.title("Corrente (I) vs Tensão de aceleração (VCG)")
plt.xlabel("Vcg [V]")
plt.ylabel("I [μA]")
plt.xlim(15, 30.8)
plt.grid(True)
plt.tight_layout()
plt.savefig("grafico_franck_hertz.png", dpi=300)  # Salva em alta resolução

plt.show()
```
